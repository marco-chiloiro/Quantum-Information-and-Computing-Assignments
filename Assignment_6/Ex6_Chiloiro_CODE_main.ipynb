```python
# plot given N and d, of the coefficients size vs separable and non-separable

N = np.linspace(1, 10, 10, dtype=int)
d = np.array([2,3,4])

plt.figure(figsize=(10, 10))
for d_val in d:
    separable_size = N*(d_val-1)
    nonsep_size = d_val**N-1
    # two different plots for separable and non-separable
    plt.subplot(2, 1, 1)
    plt.plot(N, separable_size, '--', label='d={}'.format(d_val))
    plt.scatter(N, separable_size)
    plt.xlabel('N', fontsize=14)
    plt.ylabel('Coefficients size', fontsize=14)
    plt.title('Separable', fontsize=18)
    plt.legend(fontsize=14)
    plt.xticks(N,fontsize=14)
    plt.yticks(fontsize=14)
    plt.subplot(2, 1, 2)
    plt.plot(N, nonsep_size, '--', label='d={}'.format(d_val))
    plt.scatter(N, nonsep_size)
    plt.xlabel('N', fontsize=14)
    plt.ylabel('Coefficients size', fontsize=14)
    plt.title('Non-separable', fontsize=18)
    plt.legend(fontsize=14)
    plt.xticks(N,fontsize=14)
    plt.yticks(fontsize=14)
plt.tight_layout()
plt.show()
```
