### Time-Dependent Quantum Harmonic Oscillator

```python
from Ex5_Chiloiro_CODE_myfunctions import *
```

#### Parameters

```python
# space discretization
xlow, xhigh = -10., 10.
L = xhigh-xlow
Ns = 2**12
x = np.linspace(xlow,xhigh, Ns)

# time discretization
tmax = 100.
Nt = 2001
t = np.linspace(0, tmax, Nt)
dt = t[1]-t[0]

# initial state
state_0 = eigenstate(x, 0)+1j*0

# number of frames for gifs animation
num_frames=100

# T_osc values list (>>>dt, >>dt, >dt)
T_osc = [100, 25, 5]
```

#### Results

```python
gif_names = ['slow.gif', 'medium.gif', 'fast.gif']

for jj in range(3):
    States, Pot = split_step_method(state_0,
                               [xlow,xhigh],
                               Ns,
                               [0., tmax],
                               Nt,
                               td_harmonic_potential,
                               num_frms=num_frames, 
                               T=T_osc[jj])

    Amplitudes = np.abs(States)**2

    # final time step plot and save gif
    create_and_save_animation(States, Amplitudes, Pot, x,num_frames, filename=gif_names[jj])
```
